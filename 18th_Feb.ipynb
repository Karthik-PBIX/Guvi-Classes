```python
import matplotlib.pyplot as plt
```

```python
x = [2,3,4,4.4,5,5.2,4.2,6,8,5,6,2,3,4]
y = [30,42,45,56,42,45,48,60,80,56,65,45,12,20]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from scipy import stats
```

```python
slope, intercept, r, p, stderr = stats.linregress(x,y)
```

```python
slope, intercept
```

```python
def myfunc(x, slope, intercept):
  return x*slope + intercept
```

```python
new_y = []
for x_value in x:
  new_y.append(myfunc(x_value, slope, intercept))
```

```python
y, new_y
```

```python
plt.scatter(x,y), plt.plot(x, new_y)
plt.show()
```
